# **SIMPLE NEURAL NETWORK**

```python
#Import libraries
# To build neural networks
import tensorflow as tf
# To work with numperic operations
import numpy as np
# It  is a high-level API that makes building neural networks simple
from tensorflow import keras
```

```python
# Define the simple neural network
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
# Compile the neural network
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# Try
model.compile(optimizer='sgd', loss='mean_squared_error', metrics=['mae'])
```

```python
# Provide the data
x = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
y = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

```python
# Train the data
model.fit(x, y, epochs=500)
```

```python
#  It  automatically stops training when the model is not improving
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=5)
model.fit(x, y, epochs=1000, callbacks=[callback])
```

```python
# Predict the output
print(model.predict(np.array([10.0])))
```
