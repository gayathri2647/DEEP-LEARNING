#**Binary Classification with 1 hidden layer**

```python
#Importing libraries
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Define the model
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(5,)))
model.add(Dense(1, activation='sigmoid'))
```

```python
#Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#Print the model summary
model.summary()
```

#**Multi class Classification with 1 hidden layer**

```python
# Importing libraries
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
# Example dataset
X = np.random.rand(100, 10) # 100 samples, 10 features
y = np.random.randint(0, 4, 100) # 100 labels, 4 classes (0,1,2,3)
```

```python
# One-hot encoding
y = to_categorical(y, num_classes=4)
```

```python
# Define model
model = Sequential()
model.add(Dense(16, activation='relu', input_shape=(10,))) # Hidden layer 1
model.add(Dense(8, activation='relu')) # Hidden layer 2
model.add(Dense(4, activation='softmax')) # 4 neurons in output layer to classify 4 classes
```

```python
# Compile the model
model.compile( optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X, y, epochs=20, batch_size=8)
```

```python
# Evaluate the model
loss, acc = model.evaluate(X, y)
print("Accuracy:", acc)
```

#**Regression with 1 hidden layer and tanh as an activation function**

```python
# Importing libraries
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```python
X = np.random.rand(100, 3)
input_dim = X.shape[1]   # Number of input features
```

```python
# Define model
model = Sequential([
Dense(64, activation='tanh', input_shape=(input_dim,)),
Dense(32, activation='tanh'),
Dense(1)]) # Output layer (linear activation by default)
```

```python
# Compile model
model.compile(loss='mean_squared_error', optimizer='adam')
```
